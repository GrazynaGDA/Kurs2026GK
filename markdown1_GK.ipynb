```python
# Jak napisać funkcję fizzbuzz, która:
# przyjmuje liczbę x jako dane wejściowe.
# zwraca „Fizz”, jeśli liczba jest podzielna przez 3.
# zwraca „Buzz” jeśli liczba jest podzielna przez 5.
# zwraca „FizzBuzz” jeśli liczba jest podzielna zarówno przez 3 jak i 5.

def is_divisible(x, n = 3):
    return x % n == 0

def fizzbuzz(x):
    if is_divisible(x) and is_divisible(x,5):
        return 'FizzBuzz'
    else:
        if is_divisible(x,5):
            return 'Buzz'
        else:
            if is_divisible(x) :
                return 'Fizz'
            else:
                return 'nie jest podzielna przez 3 i nie jest podzielna przez 5'

def drukujFizzBuzz(x):            
    print(x,fizzbuzz(x))   

drukujFizzBuzz(5)  
drukujFizzBuzz(15)  
drukujFizzBuzz(125)  
drukujFizzBuzz(16)  
drukujFizzBuzz(1)  
```

```python
# Jak napisać funkcję o nazwie contains_letter, która:
# Pobiera zarówno słowo jak i literę jako dane wejściowe.
# Sprawdza, czy słowo zawiera tę literę.
# Jeśli tak, wypisuje „Tak”.
# W przeciwnym razie, wypisuje „Nie”.


def contains_letter(tekst, litera):
    i = 0
    jest = 'Nie'

    while i < len(tekst) - 1 and jest == 'Nie':
        if tekst[i] == litera:
            jest = 'Tak'
        i += 1
    return jest

def drukucontyains(x,y):            
    print(x,y,contains_letter(x,y))   


drukucontyains('As', 'A')
drukucontyains('Korale', 'A')
drukucontyains('Salut', 'A')
drukucontyains('SALUT', 'A')
drukucontyains('KASA', 'S')
```
